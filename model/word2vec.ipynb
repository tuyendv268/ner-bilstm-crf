```
from gensim.models import Word2Vec 
import os
```

```
def load_data(path):
    print("Loading data..................")
    f = open(path, "r",encoding="utf-8")
    lines = f.readlines()

    datas = []
    datas_temp = []

    for line in lines:
        temp = line.replace("\n","").strip().split("\t")
        if(line == "\n"):
            datas.append(datas_temp)
            datas_temp = []
        else:
            datas_temp.append(temp[0].replace("_"," "))
    print("Load successful...............")
    return datas
```

```
datas_1 = load_data("/content/drive/MyDrive/ner/resources/data/ner_test_phonlp.txt")
datas_2 = load_data("/content/drive/MyDrive/ner/resources/data/ner_train_phonlp.txt")
datas_3 = load_data("/content/drive/MyDrive/ner/resources/data/ner_valid_phonlp.txt")
```

```
datas = datas_1 + datas_2 + datas_3
```

```
w2v_model = Word2Vec(min_count=1,
                     window=3,
                     size=64,
                     sample=6e-5, 
                     alpha=0.03, 
                     min_alpha=0.0007)
```

```
w2v_model.build_vocab(datas)
```

```
w2v_model.train(datas, total_examples=w2v_model.corpus_count, epochs=30, report_delay=1)
```

```
w2v_model.save("embeddings_vlsp_capt_64.model")
```
