```python
datas = f_in.readlines()
```

```python
def remove_punct(line):
    punc = '''!()-[]{};:'"”“\,<>./?@#$%^&*~'''
    for ele in punc:
        if ele in punc:
            line = line.replace(ele,"")
    return line
def remove_space(line):
    lines = line.split(" ")
    out = []
    for word in lines:
        if len(word) != 0 :
            out.append(word)
    return " ".join(out)
```

```python
for data in datas:
    data = data.replace("\n","")
    data = remove_punct(data)
    data = remove_space(data).strip()
    if len(data) != 0:
        data = data + "\n"
        f_out.write(data)
```

```python
from underthesea import sent_tokenize, word_tokenize
import os
import re
```

```python
raw_path = "/content/drive/MyDrive/nlp/data/data4ner/vocab/raw"
output_path = "/content/drive/MyDrive/nlp/data/data4ner/processed"
```

```python
lines = []
for file in os.listdir(path=raw_path):
    print(file)
    with open(os.path.join(raw_path,file),"r", encoding="utf-8") as txt:
        data = txt.readlines()
        lines = lines + data
```

```python
len(lines)
```

```python
def remove_punct(line):
    line = re.sub(r'[\!\"\#\$\%\&\'\(\)\*\+\,\-\.\/\:\;\<\=\>\?\@\[\\\]\^\_\`\{\|\}\~0-9]'," ",line)
    return line
def remove_space(line):
    lines = line.split(" ")
    out = []
    for word in lines:
        if len(word) != 0 and not word.isdigit():
            out.append(word)
    return " ".join(out)
def tokenize(data):
    output = []
    temp = word_tokenize(data)
    for word in temp:
      output.append(word.replace(" ","_"))
    return " ".join(output)
def data_preprocessing(data):
    data = data.replace("\n"," ")
    data = remove_punct(data)
    data = remove_space(data).strip()
    data = tokenize(data)
    return data
```

```python
temp = open(os.path.join(output_path,str(0)+".txt"), "w", encoding="utf-8")
for index in range(len(lines)):
  if index % 50000 == 0:
    temp.close()
    print(index)
    temp = open(os.path.join(output_path,str(index)+".txt"), "w", encoding="utf-8")
  line = data_preprocessing(lines[index])
  if len(line) != 0:
    temp.write(line+"\n")
```
